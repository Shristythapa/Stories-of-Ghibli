```python
def get_track_id(track_uri):
    # Split the track URI by ':'
    uri_parts = track_uri.split(':')
    # The track ID is usually the third part of the URI
    if len(uri_parts) >= 3:
        track_id = uri_parts[2]
        return track_id
    else:
        return None
```

```python
def getTrackFeature(track_id):
  # Spotify API endpoint for getting audio features
  audio_features_url = f'https://api.spotify.com/v1/audio-features/{track_id}'

  # Set the headers for the audio features request
  audio_features_headers = {
      'Authorization': f'Bearer {api_token}',
  }

  # Request audio features
  audio_features_response = requests.get(audio_features_url, headers=audio_features_headers)
  audio_features_data = audio_features_response.json()

  # Print the audio features data
  print(audio_features_data)
  return audio_features_data
```

```python
def getTrack(track_id):
  track_url = f'https://api.spotify.com/v1/tracks/{track_id}'
  audio_features_headers = {
      'Authorization': f'Bearer {api_token}',
  }
  track_response = requests.get(track_url, headers=audio_features_headers)
  track_response_data = track_response.json()
  print(track_response_data)
  return {
      'name':track_response_data['name'],
      'popularity':track_response_data['popularity'],
      'imageUrl':track_response_data['album']['images'][0]['url']
  }
```

```python
import time
import csv

playlist_with_feature = []
processed_items = 0

# Define rate limit parameters
max_requests = 20
time_limit = 30  # in seconds
start_time = time.time()
requests_made = 0

# Define the file path for CSV
csv_file_path = '/content/drive/MyDrive/ghibli_music_data.csv'

# Open the CSV file for writing in append mode
with open(csv_file_path, mode='w', newline='') as csv_file:
    writer = None  # Writer will be initialized later

    # Iterate over each playlist
    for id in ids:
        features = getTrackFeature(id)

        try:
            name = getTrack(id)
        except KeyError:
            break  # Handle missing track

        print(name)
        print(features)

        # Combine track data and features
        track_data = {
            'track_id': id,
            ** name,
            **features
        }

        # Initialize the writer on the first iteration
        if writer is None:
            # Create CSV writer and write the header
            writer = csv.DictWriter(csv_file, fieldnames=track_data.keys())
            writer.writeheader()

        # Write the track data to the CSV
        writer.writerow(track_data)

        # Update counters
        requests_made += 1
        processed_items += 1

        # Rate limit logic
        if requests_made >= max_requests:
            elapsed_time = time.time() - start_time
            if elapsed_time < time_limit:
                time.sleep(time_limit - elapsed_time)
            # Reset counters
            start_time = time.time()
            requests_made = 0

print("Total items processed:", processed_items)
print(f"Playlist data saved to {csv_file_path}")
```
